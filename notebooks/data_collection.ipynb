```python
import fastf1
import pandas as pd
```

```python
fastf1.Cache.enable_cache('../cache')
```

```python
quali = fastf1.get_session(2025, 'Canada', 'Q')
quali.load()

quali_results = quali.results
quali_results[['FullName', 'Position']]
```

```python
session = fastf1.get_session(2025, 'Canada', 'R')
session.load()
```

```python
results = session.results
results.head()
```

```python
fp2 = fastf1.get_session(2025, 'Canada', 'FP2')
fp2.load()
```

```python
laps = fp2.laps
laps.head()
```

```python
best_laps = laps.groupby('Driver')['LapTime'].min().reset_index()
best_laps
```

```python
# Race results
race_df = results[['FullName', 'Position']].copy()

# Qualifying results
quali_df = quali_results[['FullName', 'Position']].copy()
quali_df.rename(columns={'Position': 'QualiPosition'}, inplace=True)

# Practice pace
best_laps.rename(columns={
    'Driver': 'Abbreviation',
    'LapTime': 'BestFP2Lap'
}, inplace=True)

# Driver abbreviation mapping
driver_map = quali_results[['FullName', 'Abbreviation']]

# Merge practice pace with names
pace_df = pd.merge(driver_map, best_laps, on='Abbreviation')

# Merge all datasets
dataset = pd.merge(race_df, quali_df, on='FullName')
dataset = pd.merge(dataset, pace_df[['FullName', 'BestFP2Lap']], on='FullName')

dataset.head()
```

```python
dataset['BestFP2LapSeconds'] = dataset['BestFP2Lap'].dt.total_seconds()

dataset.head()
```

```python
final_dataset = dataset[['QualiPosition', 'BestFP2LapSeconds', 'Position']]

final_dataset.head()
```

```python
final_dataset.to_csv('../data/canada_2025_dataset.csv', index=False)
```
